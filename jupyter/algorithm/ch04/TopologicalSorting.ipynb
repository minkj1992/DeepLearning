```python
class TopologicalSorting:

    def __init__(self,graph,ch_n=0):
        
#         graph가 linked_list(dict안에 set으로 linkedlist를 구현)기반
        if ch_n ==0:
            self.graph = graph
#         graph가 matrix기반
        else:
            pass
#             self.graph = 
            
        
        
    def dfs_base(self, start):
        visited = []
        stack = [start]
        
        while stack:
#             print("visited:"+ str(visited))
#             print("Stack:" + str(stack))
#             print("")
        
            n = stack.pop()
            if n not in visited:
                visited.append(n)
                stack += graph[n] - set(visited)
#         print(graph["A"])
        return list(reversed(visited))


    def sourceRemoval_list(self):
        count_dict = {}
        count = [0]*len(self.graph)
        
        for i,v in enumerate(self.graph.keys()):
            count_dict[v] = i
        
        for i in graph:
#             비어있지 않다면
            for j in self.graph[i]:
                count[count_dict[j]] += 1
        
        return self.graph, count
                
        
        
```
