```python
import feedparser

feed = feedparser.parse("https://www.coindesk.com/feed/")
# 파이썬 dict 객체로 저장 되어있음.
# entries 키안에 원하는 coindesk의 새글들에 대한 정보가 담겨있다.
# entry는 feedparser 자체 구조인듯
feed
```

```python
# coindesk의 feed가 제공하는 new post 정보는 총 20개이다.
len(feed['entries'])
```

```python
#  url 들을 가져와서 저장시켜줄 객체
coindesk_urls = []

for entry in feed['entries']:
    coindesk_urls.append(entry['link']) #string append 한다.
    
```

```python
coindesk_urls
```

```python
# 파일로 저장하기

for url in coindesk_urls:
    with open("coindesk_news.txt",'a', encoding="utf8") as f:
        f.write(url+"\n")
        f.close()
```
